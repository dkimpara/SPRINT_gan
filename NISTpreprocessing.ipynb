```python
import pandas as pd
data = pd.read_csv("USA3.csv")
data.info()
data.head()
```

```python
data = data.drop(["YEAR","RACED", "HISPAND","BPLD","MBPLD","FBPLD","MTONGUED","SPANNAME","HISPRULE","HIGRADED",
                 "EDUCD","EMPSTATD","OCC","OCC1950","IND1950","CLASSWKRD","HISTID"], axis = 1)
data.info()
```

```python
for label in list(data):
    data[label] = data[label]/data[label].max()
data.head()
```

```python
d_array = data.values
X_array = d_array[1:][:]
Y_array = d_array[0][:]
```

```python
import pickle as pkl
pkl.dump(X_array, open('X_processed.pkl', 'wb'))
pkl.dump(Y_array, open('y_processed.pkl', 'wb'))
```
